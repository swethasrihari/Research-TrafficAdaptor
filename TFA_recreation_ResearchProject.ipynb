```python
import os
path = "/content/drive/MyDrive/Colab Notebooks/TrafficAdaptor/000/Trajectory"
dir_list = os.listdir(path)
print(dir_list)
```

```python
files = []
for file in dir_list:
  file_path =f'/content/drive/MyDrive/Colab Notebooks/TrafficAdaptor/000/Trajectory/{file}'
  with open(file_path, "r") as file:
        plt_content = file.readlines()
  files.append(plt_content)
print(files[0])
```

```python
mult_lat = []
mult_lon = []
mult_latLon = []

for lines in files:
  if len(lines) >= 7:  # Check if there are at least 7 lines in the file
    lat = []
    lon = []
    latLon = []
    for i in range(6,len(lines)):
      lat.append(lines[i].partition(',')[0])
      # Find the indices of the first and second ','
      first_comma_index = lines[i].index(',')
      second_comma_index = lines[i].index(',', first_comma_index + 1)

      # Extract the substring
      lon.append(lines[i][first_comma_index + 1:second_comma_index])

      latLon.append([float(lines[i].partition(',')[0]), float(lines[i][first_comma_index + 1:second_comma_index])])
    mult_lat.append(lat)
    mult_lon.append(lon)
    mult_latLon.append(latLon)
```

```python

print(mult_latLon[0])
```

```python

from sklearn.cluster import KMeans

# Convert mult_latLon to a flat list of coordinates
coordinates = [coord for sublist in mult_latLon for coord in sublist]

# Define the number of clusters
num_clusters = 2  # Adjust the number of clusters as needed

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(coordinates)
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Separate clustered points by cluster label
clustered_points = [[] for _ in range(num_clusters)]
for i, label in enumerate(labels):
    clustered_points[label].append(coordinates[i])

# Print the coordinates in each cluster
#for i, cluster in enumerate(clustered_points):
#   print(f"Cluster {i+1}:", cluster)

# Print cluster centers
print("Cluster Centers:", cluster_centers)
```

```python
import matplotlib.pyplot as plt

# Extract coordinates of each cluster
clustered_latitudes = [[] for _ in range(num_clusters)]
clustered_longitudes = [[] for _ in range(num_clusters)]
for i, cluster in enumerate(clustered_points):
    for point in cluster:
        clustered_latitudes[i].append(point[0])  # Latitude
        clustered_longitudes[i].append(point[1])  # Longitude

# Plot clustered points
plt.figure(figsize=(10, 8))
for i in range(num_clusters):
    plt.scatter(clustered_longitudes[i], clustered_latitudes[i], label=f"Cluster {i+1}")

# Plot cluster centers
plt.scatter(cluster_centers[:, 1], cluster_centers[:, 0], color='black', marker='x', label='Cluster Centers')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clustered Points')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Build Transition Matrix
import numpy as np

# Initialize transition matrix
transition_matrix = np.zeros((num_clusters, num_clusters))

# Count transitions between clusters
for i in range(len(labels) - 1):
    current_cluster = labels[i]
    next_cluster = labels[i + 1]
    transition_matrix[current_cluster][next_cluster] += 1

# Normalize transition counts to obtain transition probabilities
transition_matrix = np.array(transition_matrix / transition_matrix.sum(axis=1, keepdims=True))
#for i in range(len(transition_matrix)):
 # for element in transition_matrix[i]:
    #print(element)


print("Transition Matrix:")
print(transition_matrix)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import pairwise_distances

def discretize_transition_matrix(transition_matrix, n_clusters):
    # Calculate distances between states using cosine distance
    distances = pairwise_distances(transition_matrix, metric='euclidean')

    # Perform k-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(distances)

    # Get cluster labels for each state
    cluster_labels = kmeans.labels_

    return cluster_labels


# Discretize transition matrix
cluster_labels = discretize_transition_matrix(transition_matrix, num_clusters)
print("Cluster labels:", cluster_labels)
```
